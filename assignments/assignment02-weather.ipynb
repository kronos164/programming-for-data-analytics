# Plotting Weather Readings

***

## Importing Matplot and Pandas

***

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Loading the data from CSV file

***

```python
# Load the data
df = pd.read_csv("weatherreadings1.csv")
```

## Converting time format from the dataframe

***

```python
# Convert the "reportEndDateTime" to the right time format
df["reportEndDateTime"] = pd.to_datetime(df["reportEndDateTime"]).dt.strftime('%H:%M')
```

## Plotting the temperature over time

***

```python
# Plot the temperature over time
plt.figure(figsize=(10, 5))
plt.plot(df["reportEndDateTime"], df["dryBulbTemperature_Celsius"], label="Temperature (Celsius)")
plt.xlabel('Time')
plt.xticks(rotation=45)
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Over Time')
plt.ylim(0, 20)  # Set the y-axis range from 0 to 20
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))  # Display y-axis as integers
plt.legend()
plt.grid(True)
# Set the font size of x-axis labels and display every 4 hour
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=len(df["reportEndDateTime"].unique()) // 4))  
fixed_times = pd.date_range(start="16:00", periods=len(df["reportEndDateTime"].unique()), freq='h').strftime('%H:%M')
plt.xticks(ticks=range(0, len(df["reportEndDateTime"]),4), labels=fixed_times[::4], fontsize=7)  

plt.show()
```
