# Pie chat of the 1000 People CSV file

***

## Import dataframes

***

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

## Import CSV file data and getting current year

***

```python
today = dt.datetime.now() # get the current date
df = pd.read_csv('people-1000.csv')
df["Date of birth"] = pd.to_datetime(df["Date of birth"])
```

## Calculate age of individuals in the data and create age groups

***

```python
age = today.year - df["Date of birth"].dt.year # calculate the age

age_groups = pd.cut(age, bins=[0, 12, 18, 30, 45, 60, 80, 120], 
                    labels=["0-12", "12-18", "18-30", "30-45", 
                            "45-60", "60-80", "81+"]) # create age groups
age_group_counts = age_groups.value_counts().sort_index()
```

## Plotting pie chart of the age distribution of people

***

```python

plt.figure(figsize=(10, 7))
age_group_counts.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Age Distribution of People')
plt.ylabel('')
plt.show()
```
