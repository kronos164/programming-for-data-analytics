```python
# create data frame with the mean wind speed for each county
df_county_mean_wdsp = df_station_datails.groupby('county')['mean_wdsp'].mean().reset_index()
```

```python
gdf = gpd.GeoDataFrame(
    df_station_datails, geometry=gpd.points_from_xy(df_station_datails.longitude, 
                                                    df_station_datails.latitude, crs='EPSG:4326')
)

# Read the map of Ireland
roi_map = gpd.read_file(directory2 + irl_map)

# Merge roi_map with df_station_datails to include mean_wdsp
roi_map = roi_map.merge(df_county_mean_wdsp[['county', 'mean_wdsp']], left_on='NAME_1', right_on='county', how='left')

# Replace NaN values with 0
roi_map['mean_wdsp'] = roi_map['mean_wdsp'].fillna(0)

# Plot the map of Ireland with a larger figure size
fig, ax = plt.subplots(figsize=(20, 20))
roi_map.plot(ax=ax, column='mean_wdsp', cmap='Blues', edgecolor="black")

# Plot the GeoDataFrame with a gradient color based on mean_wdsp
gdf.plot(ax=ax, column='mean_wdsp', cmap='Blues', legend=True, markersize=100, legend_kwds={'label': "Mean Wind Speed (knot)"})


# Add county names
for x, y, label in zip(roi_map.geometry.centroid.x, roi_map.geometry.centroid.y, roi_map.NAME_1):
    
    txt = ax.text(x, y, label, fontsize=8, ha='center', color='black', weight='bold')
    txt.set_path_effects([path_effects.Stroke(linewidth=2, foreground='white'), path_effects.Normal()])

# We can now plot our ``GeoDataFrame``.
gdf.plot(ax=ax, color="lightblue", edgecolor="black", markersize=150)

# Add path effects to the station names
for x, y, label in zip(gdf.geometry.x, gdf.geometry.y, gdf['name']):
    txt = ax.text(x, y, label, fontsize=5, ha='center', color='blue', weight='bold')
    txt.set_path_effects([path_effects.Stroke(linewidth=2, foreground='white'), path_effects.Normal()])
    
# Add a title
plt.title('Mean Wind Speed in Ireland', fontsize=20)

# Remove the axis
plt.axis('off')

# Add legend to the weather stations
legend = ax.legend(["Weather Stations with Wind Speed Parameters"], loc='lower left', frameon=True)
legend.get_frame().set_edgecolor('black')
plt.show()
```

```python
windturbines = gpd.read_file(directory4 + windfarms)
df_windturbines = windturbines.groupby('County').size().reset_index(name='n_of_turbines')

# Ensure the CRS of windturbines matches the desired CRS
windturbines = windturbines.to_crs('EPSG:4326')
gdf2 = gpd.GeoDataFrame(windturbines, crs='EPSG:4326')

# Read the map of Ireland
roi_map = gpd.read_file(directory2 + irl_map)

# Merge roi_map with df_windturbines to include n_of_turbines
roi_map = roi_map.merge(df_windturbines[['County', 'n_of_turbines']], left_on='NAME_1', right_on='County', how='left')

# Replace NaN values with 0
roi_map['n_of_turbines'] = roi_map['n_of_turbines'].fillna(0)

# Plot the map of Ireland with a larger figure size
fig, ax = plt.subplots(figsize=(20, 20))
roi_map.plot(ax=ax, column='n_of_turbines', cmap='Greens', edgecolor="black", legend=True, markersize=100, legend_kwds={'label': "No of Turbines"})

# We can now plot our ``GeoDataFrame``.
gdf2.plot(ax=ax, color="lightgreen", edgecolor="black", markersize=40)

# Add county names
for x, y, label in zip(roi_map.geometry.centroid.x, roi_map.geometry.centroid.y, roi_map.NAME_1):
    
    txt = ax.text(x, y, label, fontsize=8, ha='center', color='black', weight='bold')
    txt.set_path_effects([path_effects.Stroke(linewidth=2, foreground='white'), path_effects.Normal()])

# Add a title
plt.title('Windfarms in Ireland', fontsize=20)

# Remove the axis
plt.axis('off')

# Add legend to the weather stations
legend = ax.legend(["Wind Turbines"], loc='lower left', frameon=True)
legend.get_frame().set_edgecolor('black')
plt.show()


```
